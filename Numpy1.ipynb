```python
import numpy as np

marks=np.array([1,2,3,5,6,7,8,9,9])
print(marks)
print(type(marks))
```

```python
#3d array basically for image processing

image = np.array([[[255, 0, 0],    # pixel 1 - red
                   [0, 255, 0]],   # pixel 2 - green
                  [[0, 0, 255],    # pixel 3 - blue
                   [255, 255, 0]]]) # pixel 4 - yellow
print(image)  # (2, 2, 3)
```

```python
#shape->defines rows and the colmn

marks=np.array([[1,2,3],[5,6,7],[8,9,9]])
print(marks)
print(marks.shape)
```

```python
#0 array / NUll matrix
#5 colmns and 0 rows by default 
a=np.zeros(5)
a=np.zeros((3,4)) #3 rows and 4 colmns always in tuple
print(a)
```

```python
#identity matrix
a=np.ones(9)
print(a

m=np.ones((3,4))
print(m)
```

```python
#ARANGE()

#sequential arrangement
a=np.arange(7)
print(a)

#(start,stop,gap)
a=np.arange(1,101,7)
print(a)
```

```python
#LINSPACE()
#EQUALLIY DISTRIBUTION OF NUMBER (START,STOP,PARTICITION)
a=np.linspace(1,9,10)
print(a)
```

```python
#random integer/matric
a=np.random.randint(0,5,6)
print(a)

b = np.random.rand(3, 3)
print(b)
```

```python
#np.full() — ek specific value se bharo

a=np.full(3,7) #3=col,7 value
print(a)

#matrix
a=np.full((3,6),9)
print(a)
```
